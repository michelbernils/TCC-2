```python
col_names_ce=['id', 'data_inicio', 'Nome', 'Impacto', 'Moeda'] 
ce_2020 = pd.read_csv('calendario_economico/CE_2020_USDeBRLeCNY.csv', names=col_names_ce, parse_dates=True, infer_datetime_format=True, decimal=',', skiprows=1)
ce_2019 = pd.read_csv('calendario_economico/CE_2019_USDeBRLeCNY.csv', names=col_names_ce, decimal=',', skiprows=1)
ce_2018 = pd.read_csv('calendario_economico/CE_2018_USDeBRLeCNY.csv', names=col_names_ce, decimal=',', skiprows=1)
ce_2017 = pd.read_csv('calendario_economico/CE_2017_USDeBRLeCNY.csv', names=col_names_ce, decimal=',', skiprows=1)
ce_2016 = pd.read_csv('calendario_economico/CE_2016_USDeBRLeCNY.csv', names=col_names_ce, decimal=',', skiprows=1)

col_names_ativos=['Data', 'Abertura', 'Fechamento', 'MM_21', 'MM_66', 'MM_200'] 
dol_fut = pd.read_csv('ativos/DOLFUT_2019e2020.csv', names=col_names_ativos, decimal=',', parse_dates=True, infer_datetime_format=True, skiprows=1)
vale3 = pd.read_csv('ativos/VALE3_2019e2020.csv', names=col_names_ativos, decimal=',', skiprows=1)
ibov = pd.read_csv('ativos/IBOV_2019e2020.csv', names=col_names_ativos, decimal=',', parse_dates=True, infer_datetime_format=True, skiprows=1)
petr4 = pd.read_csv('ativos/PETR4_2019e2020.csv', names=col_names_ativos, decimal=',', skiprows=1)
```

```python
#1.

ibov['data'] =  pd.to_datetime(ibov['Data']).dt.date
dol_fut['data'] =  pd.to_datetime(dol_fut['Data']).dt.date

date1 = datetime.strptime('2020-01-06', '%Y-%m-%d')
date2 = datetime.strptime('2020-01-20', '%Y-%m-%d')

ibov['data']=pd.to_datetime(ibov['data'], format='%Y-%m-%d')
dol_fut['data']=pd.to_datetime(dol_fut['data'], format='%Y-%m-%d')

ibov_week = ibov.loc[(ibov['data']>=date1) & (ibov['data']<=date2)]
dol_fut_week = dol_fut_week.loc[(dol_fut_week['data'] >= date1) & (dol_fut_week['data'] <= date2)]

#pd.concat([ibov_week, dol_fut_week], axis=1, keys=['ibov_week', 'dol_fut_week']).corr().loc['ibov_week', 'dol_fut_week']
dol_fut_week.corrwith(ibov_week)
```

```python
#2.

ibov['data'] =  pd.to_datetime(ibov['Data']).dt.date
ibov['tempo'] = pd.to_datetime(ibov['Data']).dt.time
dol_fut['data'] =  pd.to_datetime(dol_fut['Data']).dt.date
dol_fut['tempo'] = pd.to_datetime(dol_fut['Data']).dt.time

start_date = '2020-01-06'
end_date = '2020-02-06'

ibov['data'] = pd.to_datetime(ibov['data'])
mask_ibov = (ibov['data'] >= start_date) & (ibov['data'] <= end_date)

ibov_week = ibov.loc[mask_ibov]

dol_fut['data'] = pd.to_datetime(dol_fut['data'])
mask_dol = (dol_fut['data'] >= start_date) & (dol_fut['data'] <= end_date)

dol_fut_week = dol_fut.loc[mask_dol]

print(dol_fut_week.corrwith(ibov_week))
print(pd.concat([ibov_week, dol_fut_week], axis=1, keys=['ibov_week', 'dol_fut_week']).corr().loc['ibov_week', 'dol_fut_week'])
```

```python
# 3.

ibov['data'] =  pd.to_datetime(ibov['Data']).dt.date
ibov['tempo'] = pd.to_datetime(ibov['Data']).dt.time
dol_fut['data'] =  pd.to_datetime(dol_fut['Data']).dt.date
dol_fut['tempo'] = pd.to_datetime(dol_fut['Data']).dt.time
start_date = '2020-01-06'
end_date = '2020-06-06'

# 6M IBOV
ibov['data'] = pd.to_datetime(ibov['data'])
mask_ibov_6m = (ibov['data'] >= start_date) & (ibov['data'] <= end_date)
ibov_6m = ibov.loc[mask_ibov_6m]

# 6M dol_fut
dol_fut['data'] = pd.to_datetime(dol_fut['data'])
mask_dol_6m = (dol_fut['data'] >= start_date) & (dol_fut['data'] <= end_date)
dol_fut_6m = dol_fut.loc[mask_dol_6m]

# 6M IBOVESPA / 6M DOL FUT.
pd.concat([ibov_6m, dol_fut_6m], axis=1, keys=['ibov_6m', 'dol_fut_6m']).corr().loc['ibov_6m', 'dol_fut_6m']
```

```python
#4

# 1Y IBOV
ibov['data'] =  pd.to_datetime(ibov['Data']).dt.date
ibov['tempo'] = pd.to_datetime(ibov['Data']).dt.time
dol_fut['data'] =  pd.to_datetime(dol_fut['Data']).dt.date
dol_fut['tempo'] = pd.to_datetime(dol_fut['Data']).dt.time

start_date = '2020-01-06'
end_date = '2020-12-31'

ibov['data'] = pd.to_datetime(ibov['data'])
mask = (ibov['data'] >= start_date) & (ibov['data'] <= end_date)
ibov_1y = ibov.loc[mask]

dol_fut['data'] = pd.to_datetime(dol_fut['data'])
mask = (dol_fut['data'] >= start_date) & (dol_fut['data'] <= end_date)
dol_fut_1y = dol_fut.loc[mask]



# 1Y IBOVESPA / ALL DOL_FUT.
pd.concat([ibov_1y, dol_fut_1y], axis=1, keys=['ibov_1y', 'dol_fut_1y']).corr().loc['ibov_1y', 'dol_fut_1y']
```
